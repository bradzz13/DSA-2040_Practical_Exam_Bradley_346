# Association rule 


```python
#import neccessary libraries
import pandas as pd
import random
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt
```

 Generate Synthetic Transaction Data

```python
# The exam asks for ~20-50 transactions with items like milk, bread, etc.
random.seed(42) # Fixed seed for reproducibility

items_pool = ['milk', 'bread', 'butter', 'eggs', 'cheese', 'yogurt', 'cola', 'chips', 'soap', 'shampoo']
transactions = []

for i in range(50):
    # Random basket size between 3 and 7 items
    basket_size = random.randint(3, 7)
    basket = random.sample(items_pool, basket_size)
    
    # Manually inject a pattern to ensure we get interesting rules
    # Pattern: If someone buys 'milk', they also buy 'bread' (80% chance)
    if 'milk' in basket and 'bread' not in basket:
        if random.random() < 0.8:
            basket.append('bread')
            
    transactions.append(basket)

print(f"Generated {len(transactions)} transactions.")
```

Preprocessing

```python
# Convert list of lists into a One-Hot Encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_basket = pd.DataFrame(te_ary, columns=te.columns_)
print("One-Hot Encoded DataFrame:")
print(df_basket.head())
```

Apriori Algorithm

```python
# Instructions: min_support = 0.2 (Item must appear in 20% of transactions)
frequent_itemsets = apriori(df_basket, min_support=0.2, use_colnames=True)
```

Generate Association rules

```python
# Instructions: min_confidence = 0.5 (Rule must be correct 50% of the time)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```

 Display Results

```python
# Instructions: Sort by 'lift' and show top 5
top_rules = rules.sort_values('lift', ascending=False).head(5)

print("\n--- Top 5 Association Rules ---")
print(top_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
